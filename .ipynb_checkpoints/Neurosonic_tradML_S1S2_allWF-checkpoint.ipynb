```python
import os
import librosa
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from scipy.signal import firwin, filtfilt
from sklearn.pipeline import make_pipeline


import matplotlib.pyplot as plt
import seaborn as sns

from glob import glob

sample_rate = 44100
downsample_factor = 10
new_sample_rate = sample_rate // downsample_factor

use_features = ['all_WF', 's1s2', 'all_WF_s1s2'][1]
print(use_features)
seed = 68
```

```python
# define data directory
base_dir = os.getcwd()
data_dir = os.path.join(base_dir, 'Data')
```

```python
# load all files 
def load_category_data(category_path, label):
    files = glob(os.path.join(category_path, '*.wav'))
    data = []
    
    for file_path in files:
        y, sr = librosa.load(file_path, sr=None)
        data.append({
            'filename': os.path.basename(file_path),
            'waveform': y,
            'sample_rate': sr,
            'label': label,
            'length': len(y)
        })
    
    return pd.DataFrame(data)

#  create separate dataFrames
artifact_df = load_category_data(os.path.join(data_dir, 'artifact'), 'artifact')
murmur_df = load_category_data(os.path.join(data_dir, 'murmur'), 'murmur')
normal_df = load_category_data(os.path.join(data_dir, 'normal'), 'normal')

df = pd.concat([artifact_df, murmur_df, normal_df], ignore_index=True)
df['filename'] = df['filename'].str.replace('.wav', '', regex=False)
df.head()
```

```python
len(df)
```

```python
def anti_alias_and_downsample(waveform, orig_sr, factor=10):
    
    new_sr = orig_sr // factor
    cutoff = 0.8 * (0.5 * new_sr)    #  cutoff at 80% of Nyquist after downsampling
    nyquist = 0.5 * orig_sr
    norm_cutoff = cutoff / nyquist
    fir_coeff = firwin(numtaps=101, cutoff=norm_cutoff)
    filtered = filtfilt(fir_coeff, 1.0, waveform)    
    downsampled = filtered[::factor]
    
    return downsampled

df['waveform_ds'] = df.apply(lambda row: anti_alias_and_downsample(row['waveform'], row['sample_rate']), axis=1)
df['waveform_ds_length'] = df['waveform_ds'].apply(len)
```

```python
def bandpass_filter_waveform(waveform, sr, lowcut = 25, highcut = 450, numtaps=201):

    nyquist = 0.5 * sr
    low = lowcut / nyquist
    high = highcut / nyquist

    # FIR BP filter
    fir_coeff = firwin(numtaps=numtaps, cutoff=[low, high], pass_zero=False)

    # zero phase filtering
    filtered = filtfilt(fir_coeff, 1.0, waveform)

    return filtered
    
df['waveform_ds_filtered'] = df.apply(lambda row: bandpass_filter_waveform(row['waveform_ds'], sr=new_sample_rate), axis=1)
```

```python
def extract_features_from_filtered_waveform(y, sr):
    # Time domain features
    rms = np.mean(librosa.feature.rms(y=y))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y))
    peak_amp = np.max(np.abs(y))

    # Frequency domain features
    psd = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
    bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
    rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))

    # MFCCs
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfccs, axis=1)

    # Spectral entropy
    S = np.abs(librosa.stft(y))
    spectral_entropy = -np.sum(S * np.log(S + 1e-10)) / S.shape[1]

    # Aggregate and return
    features = [rms, zcr, peak_amp, psd, bandwidth, rolloff, spectral_entropy] + list(mfcc_mean)
    return features
```

```python
# All waveform based features
feature_names = [
    'rms', 'zcr', 'peak_amp',
    'centroid', 'bandwidth', 'rolloff',
    'spectral_entropy'
] + [f'mfcc_{i+1}' for i in range(13)]

feature_data = []

for _, row in df.iterrows():
    y = row['waveform_ds_filtered']
    sr = 4410  # Downsampled sampling rate
    features = extract_features_from_filtered_waveform(y, sr)
    
    feature_data.append(features + [row['label'], row['filename']])

columns = feature_names + ['label', 'filename']

df_allWF_features = pd.DataFrame(feature_data, columns=columns)
df_allWF_features.head()
```

```python
# load S1 and S2 based features
df_s1s2_features = pd.read_csv('s1s2_features.csv')
# df_s1s2_features.drop('mean_rr', axis=1, inplace=True)
df_s1s2_features.head()
```

```python
# feature selection

df_allWF_features['filename'] = df_allWF_features['filename'].astype(str)
df_s1s2_features['filename'] = df_s1s2_features['filename'].astype(str)

if use_features == 'all_WF':
    df_features = df_allWF_features
elif use_features == 's1s2':
    df_features = df_s1s2_features
elif use_features == 'all_WF_s1s2': 
    # merge both features
    df_features = pd.merge(df_allWF_features, df_s1s2_features, on='filename', suffixes=('_td', '_mfcc'))
    df_features = df_features.drop(columns=['label_td']) 
    df_features = df_features.rename(columns={'label_mfcc': 'label'})  
else:
    raise ExceptionType("Need to specify feature set")

df_features.head()
```

```python
# make features X and labels set y
X = df_features.drop(columns=['label', 'filename'])
y = df_features['label']
```

```python
# Train-test split + feature normalization
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state = seed)
filenames = df_features['filename']
filenames_train, filenames_test = train_test_split(filenames, test_size=0.2, stratify=y, random_state=seed)


# scaler = StandardScaler()
# X_train = scaler.fit_transform(X_train)
# X_test = scaler.transform(X_test)
```

```python
# Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, class_weight='balanced'),
    "SVM": SVC(probability=True),
    "Random Forest": RandomForestClassifier(random_state = seed),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier()
}
```

```python
# Training & evaluation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state = seed) #kfold ensures similar propotion of classes as in complete dataset in each fold
trained_pipelines = {}

for name, model in models.items():
    print(f"\nModel: {name}")
    
    # Train model on training set (with correct scaling)
    pipeline = make_pipeline(StandardScaler(), model)
    pipeline.fit(X_train, y_train)

    trained_pipelines[name] = pipeline

    # Evaluate on test set (properly scaled via pipeline)
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    print(f"Train Accuracy: {accuracy_score(y_train, y_train_pred):.4f}")
    print(classification_report(y_test, y_test_pred))

    # ROC-AUC
    y_proba = pipeline.predict_proba(X_test) if hasattr(pipeline, "predict_proba") else None
    if y_proba is not None:
        y_test_bin = pd.get_dummies(y_test)
        y_score = pd.DataFrame(y_proba, columns=pipeline.classes_)
        roc_auc = roc_auc_score(y_test_bin, y_score, average='macro', multi_class='ovr')
        print(f"ROC-AUC: {roc_auc:.4f}")

    # Cross-validation on training set with pipeline
    scores = cross_val_score(
        make_pipeline(StandardScaler(), model),  # new pipeline for CV
        X_train, y_train,
        cv=cv, scoring='accuracy'
    )
    print(f"Cross-Validation Accuracy: {np.mean(scores):.4f}")

    # if name == "Naive Bayes":
    #     # Create a dataframe for evaluation
    #     results_df = pd.DataFrame({
    #         'filename': filenames_test.reset_index(drop=True),
    #         'true_label': y_test.reset_index(drop=True),
    #         'predicted_label': pd.Series(y_test_pred)
    #     })
    
    #     # Filter for misclassified samples
    #     misclassified = results_df[results_df['true_label'] != results_df['predicted_label']]
    
    #     print("Misclassified cases:")
    #     print(misclassified)
```

```python
def plot_all_confusion_matrices(trained_pipelines, X_test, y_test, labels):
    n_models = len(trained_pipelines)
    cols = 2
    rows = (n_models + 1) // cols

    fig, axes = plt.subplots(rows, cols, figsize=(12, rows * 5))
    axes = axes.flatten()

    for idx, (name, pipeline) in enumerate(trained_pipelines.items()):
        y_pred = pipeline.predict(X_test)
        cm = confusion_matrix(y_test, y_pred, labels=labels)

        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=labels, yticklabels=labels, ax=axes[idx])
        axes[idx].set_title(f"{name} Confusion Matrix")
        axes[idx].set_xlabel("Predicted")
        axes[idx].set_ylabel("Actual")

    # Remove unused subplots
    for j in range(idx + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()


# Define class order
labels = ['artifact', 'murmur', 'normal']

# Plot all
plot_all_confusion_matrices(trained_pipelines, X_test, y_test, labels)
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

for name, model in models.items():
    print(f"\nModel: {name}")
    
    pipeline = make_pipeline(StandardScaler(), model)  # Scaling + model
    scores = cross_val_score(pipeline, X, y, cv=cv, scoring='accuracy')
    
    print(f"Cross-Validation Accuracy: {np.mean(scores):.4f}")
```

```python
for name, model in models.items():
    print(f"\nModel: {name}")
    all_y_true = []
    all_y_pred = []

    for train_idx, val_idx in cv.split(X, y):
        # Split data
        X_train_fold, X_val_fold = X.iloc[train_idx], X.iloc[val_idx]
        y_train_fold, y_val_fold = y.iloc[train_idx], y.iloc[val_idx]

        # Create and train pipeline with scaling
        pipeline = make_pipeline(StandardScaler(), model)
        pipeline.fit(X_train_fold, y_train_fold)

        # Predict
        preds = pipeline.predict(X_val_fold)
        all_y_true.extend(y_val_fold)
        all_y_pred.extend(preds)

    # Plot confusion matrix
    cm = confusion_matrix(all_y_true, all_y_pred, labels=np.unique(y))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y))
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix: {name}')
    plt.show()
```
