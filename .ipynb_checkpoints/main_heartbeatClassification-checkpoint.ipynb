```python
import os
import numpy as np
import pandas as pd
import utils
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline


import matplotlib.pyplot as plt
import seaborn as sns


sample_rate = 44100
downsample_factor = 10
new_sample_rate = sample_rate // downsample_factor

seed = 68
```

```python
# load S1 and S2 based features
df_features = pd.read_csv('s1s2_features.csv')
df_features.head()
```

```python
# make features X and labels set y
X = df_features.drop(columns=['label', 'filename'])
y = df_features['label']
```

```python
# train-test split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state = seed)
```

```python
# models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, class_weight='balanced'),
    "SVM": SVC(probability=True, class_weight='balanced'),
    "Random Forest": RandomForestClassifier(random_state = seed, class_weight='balanced'),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier()
}
```

```python
# training & evaluation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state = seed) 
trained_pipelines = {}

for name, model in models.items():
    print(f"\nModel: {name}")
    
    # Train model on training set 
    pipeline = make_pipeline(StandardScaler(), model)
    pipeline.fit(X_train, y_train)

    trained_pipelines[name] = pipeline 

    # Evaluate on test set 
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    print(f"Train Accuracy: {accuracy_score(y_train, y_train_pred):.4f}") 
    print(classification_report(y_test, y_test_pred)) 

    # ROC-AUC
    y_proba = pipeline.predict_proba(X_test) if hasattr(pipeline, "predict_proba") else None 
    if y_proba is not None:
        y_test_bin = pd.get_dummies(y_test)
        y_score = pd.DataFrame(y_proba, columns=pipeline.classes_)
        roc_auc = roc_auc_score(y_test_bin, y_score, average='macro', multi_class='ovr')
        print(f"ROC-AUC: {roc_auc:.4f}")

    # Cross-validation on training set 
    scores = cross_val_score(
        make_pipeline(StandardScaler(), model),  
        X_train, y_train,
        cv=cv, scoring='accuracy'
    )
    print(f"Cross-Validation Accuracy: {np.mean(scores):.4f}")
```

```python
# plot confusion matrix
labels = ['artifact', 'murmur', 'normal']
utils.plot_all_confusion_matrices(trained_pipelines, X_test, y_test, labels)
```

```python
# K-fold on all the dataset
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

for name, model in models.items():
    print(f"\nModel: {name}")
    
    pipeline = make_pipeline(StandardScaler(), model) 
    scores = cross_val_score(pipeline, X, y, cv=cv, scoring='accuracy')
    
    print(f"Cross-Validation Accuracy: {np.mean(scores):.4f}")
```
